```python
from sklearn.datasets import make_regression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x,y= make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=90,random_state=23) 
 
    
```

```python
df=pd.DataFrame(x,y)
```

```python
df
```

```python
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(x,y)
```

```python
#slope m
lr.coef_
```

```python
#intercept, b
lr.intercept_
```

```python
x[0]
```

```python
plt.plot(x,lr.predict(x),color="red")
plt.scatter(x,y)
```

```python
# code for grad descent using m from above and b =0
m=lr.coef_
b=0
epocs=5
learning_rate=0.1
#y_pred=m*x + b
plt.plot(x,lr.predict(x),color='red',label="ols(ordinary least squares)")
    
for i in range(epocs):
    loss_slope= -2 * np.sum(y-m*x.ravel()-b)
    step_size=learning_rate*loss_slope
    
    y_pred=(m*x + b).reshape(4)
    plt.plot(x,y_pred,label='b=.{}'.format(b))
    plt.legend()
    b=b- step_size
#plt.plot(x,y_pred,color="green",label='grad.{}'.format("last"))
#plt.legend()
```
